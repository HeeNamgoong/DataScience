```python
meta = pd.read_csv('/kaggle/input/the-movies-dataset/movies_metadata.csv')
sample_meta = meta[["id", "original_title","original_language", "genres"]]
sample_meta = sample_meta.rename(columns={"id":'movieId', "original_title":"title","original_language":"language"})
sample_meta
```

```python
sample_meta = sample_meta.loc[sample_meta['language'] == 'en', :]
sample_meta
```

```python
sample_meta.movieId = pd.to_numeric(sample_meta.movieId)
sample_meta.movieId
```

```python
txt = sample_meta.genres[0]

myset = set()
for d in eval(txt):
    myset.add(d['name'])
print(myset)
```

```python
def str_to_set(genre):
    myset = set()
    for d in eval(genre):
        myset.add(d['name'])
    return myset

sample_meta.genres = sample_meta.genres.apply(str_to_set)

sample_meta
```

```python
keywords = pd.read_csv('/kaggle/input/the-movies-dataset/keywords.csv')
keywords
```

```python
keywords['keywords'] = keywords['keywords'].apply(str_to_set)
keywords
```

```python
keywords = keywords.rename(columns={'id' : 'movieId'})
keywords.movieId = pd.to_numeric(keywords.movieId)
keywords
```

```python
sample_meta = pd.merge(sample_meta, keywords, on='movieId', how="inner")
sample_meta
```

```python
dk = sample_meta.loc[sample_meta['title'] == 'The Dark Knight'].iloc[0]
dkr = sample_meta.loc[sample_meta['title'] == 'The Dark Knight Rises'].iloc[0]

pd.concat([dk,dkr], axis=1).T
```

```python
def jaccard_similarity(s1, s2):
    if len(s1|s2) == 0:
        return 0
    return len(s1&s2) / len(s1|s2)
```

```python
jaccard_similarity(dk.genres | dk.keywords, dkr.genres | dkr.keywords)
```

```python
ratings = pd.read_csv('/kaggle/input/the-movies-dataset/ratings_small.csv')
ratings
```

```python
ratings = pd.merge(ratings, sample_meta[['movieId', 'title']], on='movieId', how="inner")
ratings
```

```python
matrix = ratings.pivot_table(index="userId", columns="title", values="rating")
matrix
```

```python
def pearson_similartiy(a, b):
    a = a - a.mean()
    b = b - b.mean()
    denom = (np.sum(a**2) ** 0.5) * (np.sum(b**2) ** 0.5)
    
    if denom == 0:
        return 0
    else:
        return np.sum(a*b) / denom
    
```

```python
dk_rating = matrix['The Dark Knight']
pk_rating = matrix['Prom Night']

pearson_similartiy(dk_rating, pk_rating)
```
